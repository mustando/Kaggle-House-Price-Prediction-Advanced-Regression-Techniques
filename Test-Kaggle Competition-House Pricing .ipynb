## Filtering based on missing Values

```python
df_test = df_test.loc[:, df_test.isnull().mean() < .6]
df_test
```

```python
list_train_obj2=list(df_test.select_dtypes(include="object"))
list_train_num2=list(df_test.select_dtypes(exclude="object"))
```

```python
for col in list_train_obj2:
    df_test[col]=df_test[col].fillna(df_test[col].mode()[0])
for col in list_train_num2:
    df_test[col]=df_test[col].fillna(df_test[col].mean())
```
